```python
import requests
import pandas as pd

from bs4 import BeautifulSoup

url = "https://content.codecademy.com/courses/beautifulsoup/cacao/index.html"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find(id='cacaoTable')
```

```python
# Extraire toutes les lignes (rows) de la table sauf la première (en-tête)
rows = table.find_all('tr')[1:]

# Initialiser des listes pour les ratings et le pourcentage de cacao
ratings = []
cacao_percents = []

# Parcourir chaque ligne et extraire les valeurs souhaitées
for row in rows:
    columns = row.find_all('td')
    if len(columns) >= 5:
        rating = float((row.find(class_='Rating')).get_text())
        cacao_percent = float((row.find(class_='CocoaPercent')).get_text().strip('%'))
        ratings.append(rating)
        cacao_percents.append(cacao_percent)

# Affichage des résultats
print("Ratings :", ratings)
print("Cacao Percents :", cacao_percents)
```

```python
dict_cacao = {'Rating': ratings, 'CocoaPercentage': cacao_percents}
df_cacao = pd.DataFrame(dict_cacao)
df_cacao.to_csv('cacao_data.csv')
```
